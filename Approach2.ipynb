Adding date columns

```python
def build_standard_time_features(df):
    df = df.copy()
    df['day_of_week'] = df['capturedAt'].dt.dayofweek
    df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)
    df['month'] = df['capturedAt'].dt.month
    df['day_of_year'] = df['capturedAt'].dt.dayofyear
    hour = df['capturedAt'].dt.hour
    df['is_morning'] = ((hour >= 6) & (hour < 12)).astype(int)
    df['is_afternoon'] = ((hour >= 12) & (hour < 18)).astype(int)
    df['is_evening'] = ((hour >= 18) & (hour < 24)).astype(int)
    return df
```

```python
train_df = build_standard_time_features(train_df)
validation_df = build_standard_time_features(validation_df)
test__real_df = build_standard_time_features(test__real_df)
```

Response rate is mostly empty in one shop only, imputting with median as solution

```python
# =========================================================
# GLOBAL MEDIAN (FALLBACK)
# =========================================================

global_median_response_rate = (
    train_df['shop_response_rate']
    .median()
)

# =========================================================
# SHOP-LEVEL MEDIAN
# =========================================================

shop_median_response_rate = (

    train_df
    .groupby('shopId')['shop_response_rate']
    .median()
)

# =========================================================
# MAP SHOP MEDIAN
# =========================================================

train_df['shop_response_rate_shop_median'] = (
    train_df['shopId']
    .map(shop_median_response_rate)
)

# =========================================================
# FILL MISSING
# 1. shop median
# 2. fallback global median
# =========================================================

train_df['shop_response_rate'] = (

    train_df['shop_response_rate']

    .fillna(
        train_df['shop_response_rate_shop_median']
    )

    .fillna(global_median_response_rate)
)

# =========================================================
# OPTIONAL CLEANUP
# =========================================================

train_df = train_df.drop(
    columns=['shop_response_rate_shop_median']
)
```

Combining train and sample because the derivative features need to read from previous days

```python
import numpy as np
import pandas as pd

# =========================================================
# COMBINE TRAIN + sample
# =========================================================

full_df = pd.concat([train_df, sample_df], axis=0)

# preserve indices
train_idx = train_df.index
sample_idx = sample_df.index
```

```python
# =========================================================
# SORT CHRONOLOGICALLY
# VERY IMPORTANT
# =========================================================

full_df = full_df.sort_values(
    ['shopId', 'itemId', 'modelId', 'capturedAt']
)

# =========================================================
# GROUPS
# =========================================================

# fine-grained
model_group = full_df.groupby(
    ['shopId', 'itemId', 'modelId']
)

# broader fallback
item_group = full_df.groupby(
    ['shopId', 'itemId']
)
```

```python
# =========================================================
# MODEL-LEVEL SAFE LAG FEATURES
# =========================================================

# ---------------------------------------------------------
# PREVIOUS PRICES
# ---------------------------------------------------------

full_df['model_price_last_scrape'] = (
    model_group['price'].shift(1)
)

full_df['model_price_lag_2'] = (
    model_group['price'].shift(2)
)

# ---------------------------------------------------------
# HISTORY INDICATORS
# VERY IMPORTANT
# ---------------------------------------------------------

# has at least 1 previous observation
full_df['model_has_price_history'] = (
    full_df['model_price_last_scrape']
    .notna()
    .astype(int)
)

# has at least 2 previous observations
full_df['model_has_lag_2_history'] = (
    full_df['model_price_lag_2']
    .notna()
    .astype(int)
)

# ---------------------------------------------------------
# SAFE FILLING
# ---------------------------------------------------------

# fill AFTER creating indicators
full_df['model_price_last_scrape'] = (
    full_df['model_price_last_scrape']
    .fillna(0)
)

full_df['model_price_lag_2'] = (
    full_df['model_price_lag_2']
    .fillna(0)
)

# =========================================================
# TIME FEATURES
# =========================================================

full_df['model_days_since_last_scrape'] = (
    full_df['capturedAt']
    - model_group['capturedAt'].shift(1)
).dt.days

# ---------------------------------------------------------
# HISTORY INDICATOR
# ---------------------------------------------------------

full_df['model_has_time_history'] = (
    full_df['model_days_since_last_scrape']
    .notna()
    .astype(int)
)

# ---------------------------------------------------------
# SAFE FILLING
# ---------------------------------------------------------

full_df['model_days_since_last_scrape'] = (
    full_df['model_days_since_last_scrape']
    .fillna(0)
    .astype(int)
)
```

```python
# =========================================================
# ITEM-LEVEL EXPANDING FEATURES (LEAKAGE SAFE)
# =========================================================

# item grouping
item_group = full_df.groupby(['shopId', 'itemId'])

# =========================================================
# HISTORICAL AVERAGE PRICE
# =========================================================

full_df['item_avg_price_hist'] = (
    item_group['price']
    .apply(
        lambda x: x.shift(1).expanding().mean()
    )
    .reset_index(level=[0,1], drop=True)
)

# availability flag
full_df['item_avg_price_hist_available'] = (
    full_df['item_avg_price_hist']
    .notna()
    .astype(int)
)

# safe fill
full_df['item_avg_price_hist'] = (
    full_df['item_avg_price_hist']
    .fillna(0)
)

# =========================================================
# HISTORICAL MIN PRICE
# =========================================================

full_df['item_price_min_hist'] = (
    item_group['price']
    .apply(
        lambda x: x.shift(1).expanding().min()
    )
    .reset_index(level=[0,1], drop=True)
)

# availability flag
full_df['item_price_min_hist_available'] = (
    full_df['item_price_min_hist']
    .notna()
    .astype(int)
)

# safe fill
full_df['item_price_min_hist'] = (
    full_df['item_price_min_hist']
    .fillna(0)
)

# =========================================================
# HISTORICAL MAX PRICE
# =========================================================

full_df['item_price_max_hist'] = (
    item_group['price']
    .apply(
        lambda x: x.shift(1).expanding().max()
    )
    .reset_index(level=[0,1], drop=True)
)

# availability flag
full_df['item_price_max_hist_available'] = (
    full_df['item_price_max_hist']
    .notna()
    .astype(int)
)

# safe fill
full_df['item_price_max_hist'] = (
    full_df['item_price_max_hist']
    .fillna(0)
)

# =========================================================
# HISTORICAL VOLATILITY
# =========================================================

full_df['item_price_volatility_hist'] = (
    item_group['price']
    .apply(
        lambda x: x.shift(1).expanding().std()
    )
    .reset_index(level=[0,1], drop=True)
)

# availability flag
full_df['item_price_volatility_hist_available'] = (
    full_df['item_price_volatility_hist']
    .notna()
    .astype(int)
)

# safe fill
full_df['item_price_volatility_hist'] = (
    full_df['item_price_volatility_hist']
    .fillna(0)
)

# =========================================================
# OPTIONAL:
# HISTORICAL OBSERVATION COUNT
# VERY USEFUL FEATURE
# =========================================================

full_df['item_hist_observation_count'] = (
    item_group.cumcount()
)

# =========================================================
# OPTIONAL:
# HAS ANY HISTORY
# =========================================================

full_df['item_has_history'] = (
    full_df['item_hist_observation_count'] > 0
).astype(int)
```

```python
# =========================================================
# MODEL-LEVEL PRICE CHANGE FEATURES (PROPER + SAFE)
# =========================================================

# ---------------------------------------------------------
# PREVIOUS PRICE CHANGE
# ---------------------------------------------------------

full_df['model_prev_price_change'] = (
    full_df['model_price_last_scrape']
    - full_df['model_price_lag_2']
)

# =========================================================
# AVAILABILITY FLAG
# need BOTH lag1 and lag2 to exist
# =========================================================

full_df['model_prev_price_change_available'] = (
    (
        full_df['model_has_price_history'] == 1
    )
    &
    (
        full_df['model_has_lag_2_history'] == 1
    )
).astype(int)

# =========================================================
# SAFE FILL
# first/second occurrences have no meaningful change
# =========================================================

full_df['model_prev_price_change'] = (
    full_df['model_prev_price_change']
    .fillna(0)
)

# =========================================================
# PERCENTAGE CHANGE
# =========================================================

full_df['model_prev_price_change_pct'] = np.where(
    full_df['model_has_lag_2_history'] == 1,
    (
        full_df['model_prev_price_change']
        /
        full_df['model_price_lag_2'].replace(0, 1)
    ) * 100,
    np.nan
)

# =========================================================
# AVAILABILITY FLAG
# =========================================================

full_df['model_prev_price_change_pct_available'] = (
    full_df['model_prev_price_change_pct']
    .notna()
    .astype(int)
)

# =========================================================
# SAFE FILL
# =========================================================

full_df['model_prev_price_change_pct'] = (
    full_df['model_prev_price_change_pct']
    .fillna(0)
)

# =========================================================
# OPTIONAL:
# ABSOLUTE CHANGE MAGNITUDE
# often very useful for tree models
# =========================================================

full_df['model_prev_price_change_abs'] = (
    full_df['model_prev_price_change']
    .abs()
)

# =========================================================
# OPTIONAL:
# PRICE DIRECTION
# =========================================================

full_df['model_price_direction'] = np.sign(
    full_df['model_prev_price_change']
)

# values:
# -1 = price decreased
#  0 = unchanged/no history
# +1 = price increased
```

```python
# =========================================================
# MODEL-LEVEL ROLLING FEATURES (LEAKAGE SAFE)
# =========================================================

# model grouping
model_group = full_df.groupby(
    ['shopId', 'itemId', 'modelId']
)

# =========================================================
# ROLLING 7-OBSERVATION AVERAGE
# =========================================================

full_df['model_price_7d_avg_hist'] = (
    model_group['price']
    .apply(
        lambda x: x.shift(1).rolling(
            window=7,
            min_periods=1
        ).mean()
    )
    .reset_index(level=[0,1,2], drop=True)
)

# availability flag
full_df['model_price_7d_avg_hist_available'] = (
    full_df['model_price_7d_avg_hist']
    .notna()
    .astype(int)
)

# safe fill
full_df['model_price_7d_avg_hist'] = (
    full_df['model_price_7d_avg_hist']
    .fillna(0)
)

# ===========
```

```python
# =========================================================
# ITEM-LEVEL LAG FEATURES
# =========================================================

item_group = full_df.groupby(['shopId', 'itemId'])

# ---------------------------------------------------------
# LAGS
# ---------------------------------------------------------

full_df['item_price_last_scrape'] = (
    item_group['price'].shift(1)
)

full_df['item_price_lag_2'] = (
    item_group['price'].shift(2)
)

# =========================================================
# HISTORY FLAGS
# =========================================================

full_df['item_has_price_history'] = (
    full_df['item_price_last_scrape']
    .notna()
    .astype(int)
)

full_df['item_has_lag_2_history'] = (
    full_df['item_price_lag_2']
    .notna()
    .astype(int)
)

# =========================================================
# SAFE FILL
# =========================================================

full_df['item_price_last_scrape'] = (
    full_df['item_price_last_scrape']
    .fillna(0)
)

full_df['item_price_lag_2'] = (
    full_df['item_price_lag_2']
    .fillna(0)
)
```

```python
# =========================================================
# ITEM-LEVEL PRICE CHANGE FEATURES (PROPER + SAFE)
# =========================================================

# ---------------------------------------------------------
# PREVIOUS PRICE CHANGE
# ---------------------------------------------------------

full_df['item_prev_price_change'] = (
    full_df['item_price_last_scrape']
    - full_df['item_price_lag_2']
)

# =========================================================
# AVAILABILITY FLAG
# need BOTH lag1 and lag2 to exist
# =========================================================

full_df['item_prev_price_change_available'] = (
    (
        full_df['item_has_price_history'] == 1
    )
    &
    (
        full_df['item_has_lag_2_history'] == 1
    )
).astype(int)

# =========================================================
# SAFE FILL
# =========================================================

full_df['item_prev_price_change'] = (
    full_df['item_prev_price_change']
    .fillna(0)
)

# =========================================================
# PERCENTAGE CHANGE
# =========================================================

full_df['item_prev_price_change_pct'] = np.where(
    full_df['item_has_lag_2_history'] == 1,
    (
        full_df['item_prev_price_change']
        /
        full_df['item_price_lag_2'].replace(0, 1)
    ) * 100,
    np.nan
)

# =========================================================
# AVAILABILITY FLAG
# =========================================================

full_df['item_prev_price_change_pct_available'] = (
    full_df['item_prev_price_change_pct']
    .notna()
    .astype(int)
)

# =========================================================
# SAFE FILL
# =========================================================

full_df['item_prev_price_change_pct'] = (
    full_df['item_prev_price_change_pct']
    .fillna(0)
)

# =========================================================
# OPTIONAL:
# ABSOLUTE CHANGE MAGNITUDE
# =========================================================

full_df['item_prev_price_change_abs'] = (
    full_df['item_prev_price_change']
    .abs()
)

# =========================================================
# OPTIONAL:
# PRICE DIRECTION
# =========================================================

full_df['item_price_direction'] = np.sign(
    full_df['item_prev_price_change']
)

# values:
# -1 = price decreased
#  0 = unchanged/no history
# +1 = price increased
```

```python
# =========================================================
# MODEL-LEVEL EXPANDING FEATURES (SAFE)
# =========================================================

model_group = full_df.groupby(
    ['shopId', 'itemId', 'modelId']
)

# =========================================================
# HISTORICAL AVG PRICE
# =========================================================

full_df['model_avg_price_hist'] = (
    model_group['price']
    .apply(
        lambda x: x.shift(1).expanding().mean()
    )
    .reset_index(level=[0,1,2], drop=True)
)

full_df['model_avg_price_hist_available'] = (
    full_df['model_avg_price_hist']
    .notna()
    .astype(int)
)

full_df['model_avg_price_hist'] = (
    full_df['model_avg_price_hist']
    .fillna(0)
)

# =========================================================
# HISTORICAL MIN PRICE
# =========================================================

full_df['model_price_min_hist'] = (
    model_group['price']
    .apply(
        lambda x: x.shift(1).expanding().min()
    )
    .reset_index(level=[0,1,2], drop=True)
)

full_df['model_price_min_hist_available'] = (
    full_df['model_price_min_hist']
    .notna()
    .astype(int)
)

full_df['model_price_min_hist'] = (
    full_df['model_price_min_hist']
    .fillna(0)
)

# =========================================================
# HISTORICAL MAX PRICE
# =========================================================

full_df['model_price_max_hist'] = (
    model_group['price']
    .apply(
        lambda x: x.shift(1).expanding().max()
    )
    .reset_index(level=[0,1,2], drop=True)
)

full_df['model_price_max_hist_available'] = (
    full_df['model_price_max_hist']
    .notna()
    .astype(int)
)

full_df['model_price_max_hist'] = (
    full_df['model_price_max_hist']
    .fillna(0)
)

# =========================================================
# HISTORICAL VOLATILITY
# =========================================================

full_df['model_price_volatility_hist'] = (
    model_group['price']
    .apply(
        lambda x: x.shift(1).expanding().std()
    )
    .reset_index(level=[0,1,2], drop=True)
)

full_df['model_price_volatility_hist_available'] = (
    full_df['model_price_volatility_hist']
    .notna()
    .astype(int)
)

full_df['model_price_volatility_hist'] = (
    full_df['model_price_volatility_hist']
    .fillna(0)
)

# =========================================================
# OPTIONAL:
# HISTORY COUNT
# =========================================================

full_df['model_hist_observation_count'] = (
    model_group.cumcount()
)

full_df['model_has_history'] = (
    full_df['model_hist_observation_count'] > 0
).astype(int)
```

```python
# =========================================================
# MODEL-LEVEL PRICE POSITION FEATURES (SAFE)
# =========================================================

# ---------------------------------------------------------
# POSITION IN HISTORICAL RANGE
# ---------------------------------------------------------

full_df['model_price_position_hist'] = np.where(

    (
        full_df['model_price_max_hist']
        - full_df['model_price_min_hist']
    ) != 0,

    (
        full_df['model_price_last_scrape']
        - full_df['model_price_min_hist']
    )
    /
    (
        full_df['model_price_max_hist']
        - full_df['model_price_min_hist']
    ),

    np.nan
)

# =========================================================
# AVAILABILITY FLAG
# requires:
# - history exists
# - min/max history exists
# =========================================================

full_df['model_price_position_hist_available'] = (
    (
        full_df['model_has_price_history'] == 1
    )
    &
    (
        full_df['model_price_min_hist_available'] == 1
    )
    &
    (
        full_df['model_price_max_hist_available'] == 1
    )
).astype(int)

# =========================================================
# SAFE FILL
# =========================================================

full_df['model_price_position_hist'] = (
    full_df['model_price_position_hist']
    .fillna(0)
)

# =========================================================
# OPTIONAL:
# OUTSIDE HISTORICAL RANGE FLAGS
# =========================================================

full_df['model_price_above_hist_max'] = (
    full_df['model_price_last_scrape']
    > full_df['model_price_max_hist']
).astype(int)

full_df['model_price_below_hist_min'] = (
    full_df['model_price_last_scrape']
    < full_df['model_price_min_hist']
).astype(int)

# =========================================================
# =========================================================
# ITEM-LEVEL PRICE POSITION FEATURES (SAFE)
# =========================================================
# =========================================================

# ---------------------------------------------------------
# POSITION IN HISTORICAL RANGE
# ---------------------------------------------------------

full_df['item_price_position_hist'] = np.where(

    (
        full_df['item_price_max_hist']
        - full_df['item_price_min_hist']
    ) != 0,

    (
        full_df['item_price_last_scrape']
        - full_df['item_price_min_hist']
    )
    /
    (
        full_df['item_price_max_hist']
        - full_df['item_price_min_hist']
    ),

    np.nan
)

# =========================================================
# AVAILABILITY FLAG
# =========================================================

full_df['item_price_position_hist_available'] = (
    (
        full_df['item_has_price_history'] == 1
    )
    &
    (
        full_df['item_price_min_hist_available'] == 1
    )
    &
    (
        full_df['item_price_max_hist_available'] == 1
    )
).astype(int)

# =========================================================
# SAFE FILL
# =========================================================

full_df['item_price_position_hist'] = (
    full_df['item_price_position_hist']
    .fillna(0)
)

# =========================================================
# OPTIONAL:
# OUTSIDE HISTORICAL RANGE FLAGS
# =========================================================

full_df['item_price_above_hist_max'] = (
    full_df['item_price_last_scrape']
    > full_df['item_price_max_hist']
).astype(int)

full_df['item_price_below_hist_min'] = (
    full_df['item_price_last_scrape']
    < full_df['item_price_min_hist']
).astype(int)
```

```python
# =========================================================
# MODEL-LEVEL TREND FEATURES (SAFE)
# =========================================================

# ---------------------------------------------------------
# TREND PER DAY
# ---------------------------------------------------------

full_df['model_price_trend_per_day'] = np.where(

    (
        full_df['model_prev_price_change_available'] == 1
    )
    &
    (
        full_df['model_has_time_history'] == 1
    )
    &
    (
        full_df['model_days_since_last_scrape'] > 0
    ),

    (
        full_df['model_prev_price_change']
        /
        full_df['model_days_since_last_scrape']
    ),

    np.nan
)

# =========================================================
# AVAILABILITY FLAG
# =========================================================

full_df['model_price_trend_per_day_available'] = (
    full_df['model_price_trend_per_day']
    .notna()
    .astype(int)
)

# =========================================================
# SAFE FILL
# =========================================================

full_df['model_price_trend_per_day'] = (
    full_df['model_price_trend_per_day']
    .fillna(0)
)

# =========================================================
# OPTIONAL:
# ABSOLUTE TREND
# =========================================================

full_df['model_price_trend_abs'] = (
    full_df['model_price_trend_per_day']
    .abs()
)

# =========================================================
# OPTIONAL:
# TREND DIRECTION
# =========================================================

full_df['model_price_trend_direction'] = np.sign(
    full_df['model_price_trend_per_day']
)
```

```python
# =========================================================
# ITEM-LEVEL TIME FEATURES
# =========================================================

item_group = full_df.groupby(['shopId', 'itemId'])

# ---------------------------------------------------------
# DAYS SINCE LAST SCRAPE
# ---------------------------------------------------------

full_df['item_days_since_last_scrape'] = (
    full_df['capturedAt']
    - item_group['capturedAt'].shift(1)
).dt.days

# =========================================================
# HISTORY FLAG
# =========================================================

full_df['item_has_time_history'] = (
    full_df['item_days_since_last_scrape']
    .notna()
    .astype(int)
)

# =========================================================
# SAFE FILL
# =========================================================

full_df['item_days_since_last_scrape'] = (
    full_df['item_days_since_last_scrape']
    .fillna(0)
    .astype(int)
)

# =========================================================
# OPTIONAL:
# DAYS SINCE FIRST SEEN
# =========================================================

full_df['item_days_since_first_seen'] = (
    full_df['capturedAt']
    - item_group['capturedAt'].transform('min')
).dt.days
```

```python
# =========================================================
# ITEM-LEVEL TREND FEATURES (SAFE)
# =========================================================

# ---------------------------------------------------------
# TREND PER DAY
# ---------------------------------------------------------

full_df['item_price_trend_per_day'] = np.where(

    (
        full_df['item_prev_price_change_available'] == 1
    )
    &
    (
        full_df['item_has_time_history'] == 1
    )
    &
    (
        full_df['item_days_since_last_scrape'] > 0
    ),

    (
        full_df['item_prev_price_change']
        /
        full_df['item_days_since_last_scrape']
    ),

    np.nan
)

# =========================================================
# AVAILABILITY FLAG
# =========================================================

full_df['item_price_trend_per_day_available'] = (
    full_df['item_price_trend_per_day']
    .notna()
    .astype(int)
)

# =========================================================
# SAFE FILL
# =========================================================

full_df['item_price_trend_per_day'] = (
    full_df['item_price_trend_per_day']
    .fillna(0)
)

# =========================================================
# OPTIONAL:
# ABSOLUTE TREND
# =========================================================

full_df['item_price_trend_abs'] = (
    full_df['item_price_trend_per_day']
    .abs()
)

# =========================================================
# OPTIONAL:
# TREND DIRECTION
# =========================================================

full_df['item_price_trend_direction'] = np.sign(
    full_df['item_price_trend_per_day']
)
```

```python
# =========================================================
# DISCOUNT FEATURES (SAFE + PRODUCTION GRADE)
# =========================================================

# ---------------------------------------------------------
# ORIGINAL PRICE AVAILABLE
# ---------------------------------------------------------

full_df['price_before_discount_available'] = (
    full_df['priceBeforeDiscount']
    .notna()
    .astype(int)
)

# ---------------------------------------------------------
# SAFE FILL
# assumption:
# missing original price = no discount
# ---------------------------------------------------------

full_df['priceBeforeDiscount'] = (
    full_df['priceBeforeDiscount']
    .fillna(full_df['price'])
)

# =========================================================
# IS ON SALE
# =========================================================

full_df['is_on_sale'] = (
    full_df['price']
    < full_df['priceBeforeDiscount']
).astype(int)

# =========================================================
# RAW DISCOUNT AMOUNT
# =========================================================

full_df['discount_amount'] = (
    full_df['priceBeforeDiscount']
    - full_df['price']
)

# =========================================================
# DISCOUNT DEPTH
# normalized discount intensity
# =========================================================

full_df['discount_depth'] = np.where(

    full_df['priceBeforeDiscount'] > 0,

    (
        full_df['discount_amount']
        /
        full_df['priceBeforeDiscount']
    ),

    np.nan
)

# availability flag
full_df['discount_depth_available'] = (
    full_df['discount_depth']
    .notna()
    .astype(int)
)

# safe fill
full_df['discount_depth'] = (
    full_df['discount_depth']
    .fillna(0)
)

# =========================================================
# EFFECTIVE DISCOUNT %
# if show_discount exists
# =========================================================

full_df['show_discount_available'] = (
    full_df['show_discount']
    .notna()
    .astype(int)
)

full_df['effective_discount_pct'] = (
    full_df['show_discount'] / 100
)

full_df['effective_discount_pct'] = (
    full_df['effective_discount_pct']
    .fillna(0)
)

# =========================================================
# OPTIONAL:
# DISCOUNT VS HISTORICAL ITEM AVG
# VERY POWERFUL FEATURE
# =========================================================

full_df['price_vs_item_hist_avg_pct'] = np.where(

    full_df['item_avg_price_hist_available'] == 1,

    (
        (
            full_df['price']
            - full_df['item_avg_price_hist']
        )
        /
        full_df['item_avg_price_hist'].replace(0, 1)
    ) * 100,

    np.nan
)

full_df['price_vs_item_hist_avg_pct_available'] = (
    full_df['price_vs_item_hist_avg_pct']
    .notna()
    .astype(int)
)

full_df['price_vs_item_hist_avg_pct'] = (
    full_df['price_vs_item_hist_avg_pct']
    .fillna(0)
)

# =========================================================
# OPTIONAL:
# SHARP DISCOUNT FLAG
# =========================================================

full_df['is_sharp_discount'] = (
    full_df['model_prev_price_change_pct'] < -20
).astype(int)

# =========================================================
# POPULARITY FEATURES
# =========================================================

# ---------------------------------------------------------
# SHOP POPULARITY
# ---------------------------------------------------------

full_df['shop_follower_count_available'] = (
    full_df['shop_follower_count']
    .notna()
    .astype(int)
)

full_df['shop_follower_count'] = (
    full_df['shop_follower_count']
    .fillna(0)
)

full_df['shop_popularity'] = np.log1p(
    full_df['shop_follower_count']
)

# =========================================================
# ITEM POPULARITY
# =========================================================

full_df['total_rating_count_available'] = (
    full_df['total_rating_count']
    .notna()
    .astype(int)
)

full_df['total_rating_count'] = (
    full_df['total_rating_count']
    .fillna(0)
)

full_df['item_popularity'] = np.log1p(
    full_df['total_rating_count']
)

# =========================================================
# REVIEW SCORE
# =========================================================

full_df['review_rating_available'] = (
    full_df['review_rating']
    .notna()
    .astype(int)
)

# fill missing with neutral score
full_df['review_rating'] = (
    full_df['review_rating']
    .fillna(0)
)

full_df['review_score_normalized'] = (
    full_df['review_rating'] / 5.0
)

# =========================================================
# OPTIONAL:
# INTERACTION FEATURES
# VERY STRONG FOR TREE MODELS
# =========================================================

full_df['discount_x_popularity'] = (
    full_df['discount_depth']
    * full_df['item_popularity']
)

full_df['discount_x_review_score'] = (
    full_df['discount_depth']
    * full_df['review_score_normalized']
)
```

```python
# =========================================================
# LOG TRANSFORM FEATURES
# =========================================================

import numpy as np

# =========================================================
# HELPER FUNCTION
# for features that can be negative
# =========================================================

def signed_log1p(x):
    return np.sign(x) * np.log1p(np.abs(x))

# =========================================================
# POSITIVE FEATURES
# =========================================================

log_features = [

    # -----------------------------------------------------
    # ITEM HISTORICAL PRICE FEATURES
    # -----------------------------------------------------

    'item_avg_price_hist',
    'item_price_min_hist',
    'item_price_max_hist',
    'item_price_volatility_hist',

    # -----------------------------------------------------
    # MODEL HISTORICAL PRICE FEATURES
    # -----------------------------------------------------

    'model_avg_price_hist',
    'model_price_min_hist',
    'model_price_max_hist',
    'model_price_volatility_hist',

    # -----------------------------------------------------
    # LAG FEATURES
    # -----------------------------------------------------

    'item_price_last_scrape',
    'item_price_lag_2',

    'model_price_last_scrape',
    'model_price_lag_2',

    # -----------------------------------------------------
    # DISCOUNT FEATURES
    # -----------------------------------------------------

    'discount_amount',
    'priceBeforeDiscount',

    # -----------------------------------------------------
    # OBSERVATION / HISTORY FEATURES
    # -----------------------------------------------------

    'item_hist_observation_count',
    'model_hist_observation_count',

    # -----------------------------------------------------
    # TIME FEATURES
    # -----------------------------------------------------

    'item_days_since_first_seen'
]

# =========================================================
# CREATE LOG FEATURES
# =========================================================

for col in log_features:

    if col in full_df.columns:

        full_df[f'{col}_log1p'] = np.log1p(
            full_df[col].clip(lower=0)
        )

# =========================================================
# SIGNED FEATURES
# =========================================================

signed_features = [

    # -----------------------------------------------------
    # PRICE CHANGES
    # -----------------------------------------------------

    'model_prev_price_change',
    'item_prev_price_change',

    # -----------------------------------------------------
    # PERCENTAGE CHANGES
    # -----------------------------------------------------

    'model_prev_price_change_pct',
    'item_prev_price_change_pct',

    # -----------------------------------------------------
    # TREND FEATURES
    # -----------------------------------------------------

    'model_price_trend_per_day',
    'item_price_trend_per_day'
]

# =========================================================
# CREATE SIGNED LOG FEATURES
# =========================================================

for col in signed_features:

    if col in full_df.columns:

        full_df[f'{col}_signed_log1p'] = (
            np.sign(full_df[col])
            *
            np.log1p(np.abs(full_df[col]))
        )

# =========================================================
# OPTIONAL:
# LOG TRANSFORM TARGET
# =========================================================

full_df['price_log1p'] = np.log1p(
    full_df['price']
)

print("Log-transformed features created successfully.")
```

```python
# =========================================================
# ITEM-LEVEL LOG-IQR OUTLIER DETECTION
# ONLY FLAG:
# - ABOVE UPPER BOUND
# - is_on_sale == 0
# =========================================================

import numpy as np
import pandas as pd

# =========================================================
# SANDBOX COPY
# =========================================================

full_df2 = full_df.copy()

# =========================================================
# PARAMETERS
# =========================================================

IQR_MULTIPLIER = 2
MIN_ITEM_COUNT = 5

# =========================================================
# TRAIN DATA ONLY
# =========================================================

train_df2 = full_df2[
    full_df2['tag2'] == 'TRAIN'
].copy()

# =========================================================
# LOG PRICE
# =========================================================

train_df2['log_price'] = np.log1p(
    train_df2['price']
)

full_df2['log_price'] = np.log1p(
    full_df2['price']
)

# =========================================================
# ITEM-LEVEL LOG-IQR STATS
# =========================================================

item_stats = (

    train_df2
    .groupby('itemId')['log_price']
    .agg(
        q1=lambda x: x.quantile(0.25),
        q3=lambda x: x.quantile(0.75),
        median='median',
        count='count'
    )
)

# =========================================================
# IQR
# =========================================================

item_stats['iqr'] = (
    item_stats['q3']
    - item_stats['q1']
)

# =========================================================
# LOWER / UPPER BOUNDS
# =========================================================

item_stats['lower_bound'] = (

    item_stats['q1']
    - IQR_MULTIPLIER * item_stats['iqr']
)

item_stats['upper_bound'] = (

    item_stats['q3']
    + IQR_MULTIPLIER * item_stats['iqr']
)

# =========================================================
# MERGE BACK
# =========================================================

full_df2 = full_df2.merge(

    item_stats[
        [
            'lower_bound',
            'upper_bound',
            'count'
        ]
    ],

    on='itemId',
    how='left'
)

# =========================================================
# OUTLIER FLAG
# ONLY:
# - ABOVE UPPER BOUND
# - NOT ON SALE
# - ENOUGH HISTORY
# =========================================================

full_df2['item_log_iqr_outlier'] = np.where(

    (
        (full_df2['count'] >= MIN_ITEM_COUNT)
        &
        (full_df2['log_price'] > full_df2['upper_bound'])
        &
        (full_df2['is_on_sale'] == 0)
    ),

    1,

    0
)

# =========================================================
# DISTANCE ABOVE UPPER BOUND
# =========================================================

full_df2['item_iqr_distance'] = np.where(

    full_df2['log_price']
    > full_df2['upper_bound'],

    full_df2['log_price']
    - full_df2['upper_bound'],

    0
)

# =========================================================
# SUMMARY
# =========================================================

print("=" * 60)

print(
    f"Detected outliers: "
    f"{full_df2['item_log_iqr_outlier'].sum():,}"
)

print("=" * 60)

# =========================================================
# VIEW EXAMPLES
# =========================================================

outlier_examples = full_df2[
    full_df2['item_log_iqr_outlier'] == 1
].copy()

# =========================================================
# SORT MOST EXTREME
# =========================================================

outlier_examples = outlier_examples.sort_values(
    by='item_iqr_distance',
    ascending=False
)

# =========================================================
# DISPLAY WITH REFERENCE PRICES
# =========================================================

outlier_examples['lower_bound_price'] = np.expm1(
    outlier_examples['lower_bound']
)

outlier_examples['upper_bound_price'] = np.expm1(
    outlier_examples['upper_bound']
)

# =========================================================
# DISPLAY COLUMNS
# =========================================================

display_cols = [

    'capturedAt',

    'shopId',
    'itemId',
    'modelId',

    # current pricing
    'price',
    'priceBeforeDiscount',

    # discount info
    'is_on_sale',
    'discount_amount',
    'discount_depth',
    'effective_discount_pct',
    'is_sharp_discount',

    # log price
    'log_price',

    # historical references
    'item_avg_price_hist',
    'item_price_min_hist',
    'item_price_max_hist',

    # IQR bounds
    'lower_bound_price',
    'upper_bound_price',

    # anomaly strength
    'item_iqr_distance',

    # history count
    'count'
]

# =========================================================
# SHOW TOP OUTLIERS
# =========================================================

display(
    outlier_examples[
        display_cols
    ].head(10)
)

# =========================================================
# SEE TEMP COLUMNS CREATED
# =========================================================

new_cols = sorted(
    list(
        set(full_df2.columns)
        -
        set(full_df.columns)
    )
)

print("\nNEW TEMPORARY COLUMNS:\n")

for col in new_cols:
    print(col)
```

```python
# =========================================================
# SORT clean_df
# =========================================================

clean_df_sorted = clean_df.sort_values(
    'capturedAt'
)

# =========================================================
# MODEL-LEVEL LATEST LOOKUP
# =========================================================

model_lookup = (

    clean_df_sorted

    .groupby(
        ['shopId', 'itemId', 'modelId'],
        as_index=False
    )

    .last()
)

# =========================================================
# ITEM-LEVEL FALLBACK LOOKUP
# =========================================================

item_lookup = (

    clean_df_sorted

    .groupby(
        ['shopId', 'itemId'],
        as_index=False
    )

    .last()
)

# =========================================================
# EXCLUDE TARGET / ID COLS
# =========================================================

exclude_cols = [

    'capturedAt',
    'shopId',
    'itemId',
    'modelId',

    'price',
    'price_log1p',
    'log_price'
]

fill_cols = [
    col for col in remaining_df2.columns
    if col not in exclude_cols
]

# =========================================================
# START COPY
# =========================================================

filled_df = remaining_df2.copy()

# =========================================================
# FILL FEATURES
# PRIORITY:
# 1. exact modelId
# 2. fallback itemId
# =========================================================

for col in fill_cols:

    # -----------------------------------------------------
    # TEMP COL NAMES
    # -----------------------------------------------------

    model_temp_col = f"__model_{col}"

    item_temp_col = f"__item_{col}"

    # -----------------------------------------------------
    # MODEL LOOKUP
    # -----------------------------------------------------

    model_col = model_lookup[
        [
            'shopId',
            'itemId',
            'modelId',
            col
        ]
    ].rename(
        columns={
            col: model_temp_col
        }
    )

    filled_df = filled_df.merge(

        model_col,

        on=['shopId', 'itemId', 'modelId'],

        how='left'
    )

    # -----------------------------------------------------
    # ITEM LOOKUP
    # -----------------------------------------------------

    item_col = item_lookup[
        [
            'shopId',
            'itemId',
            col
        ]
    ].rename(
        columns={
            col: item_temp_col
        }
    )

    filled_df = filled_df.merge(

        item_col,

        on=['shopId', 'itemId'],

        how='left'
    )

    # -----------------------------------------------------
    # FILL PRIORITY
    # current -> model -> item
    # -----------------------------------------------------

    filled_df[col] = (

        filled_df[col]

        .fillna(
            filled_df[model_temp_col]
        )

        .fillna(
            filled_df[item_temp_col]
        )
    )

    # -----------------------------------------------------
    # DROP TEMP COLS
    # -----------------------------------------------------

    filled_df = filled_df.drop(
        columns=[
            model_temp_col,
            item_temp_col
        ]
    )

# =========================================================
# CHECK
# =========================================================

print("=" * 60)
print("Remaining NaNs")
print("=" * 60)

print(
    filled_df
    .isna()
    .sum()
    .sort_values(ascending=False)
    .head(5)
)
```

```python
# =========================================================
# MULTI-DAY INFERENCE PIPELINE
# ENTITY-LEVEL DRIFT CALIBRATION
# =========================================================

import numpy as np
import pandas as pd

# =========================================================
# COPY
# =========================================================

infer_df = test__real_df.copy()

# =========================================================
# ADD ANCHOR FLAG
# =========================================================

infer_df['is_anchor'] = (
    infer_df['price'].notna()
).astype(int)

# =========================================================
# ADD ANCHOR FLAG TO CLEAN_DF
# =========================================================

clean_df['is_anchor'] = 0

# =========================================================
# ADD MISSING COLS TO infer_df
# =========================================================

for col in clean_df.columns:

    if col not in infer_df.columns:

        infer_df[col] = np.nan

# =========================================================
# MATCH EXACT COLUMN ORDER
# =========================================================

infer_df = infer_df[
    clean_df.columns
]

# =========================================================
# CONCAT
# =========================================================

combined_df = pd.concat(

    [
        clean_df,
        infer_df
    ],

    ignore_index=True
)

# =========================================================
# MARK FUTURE ROWS
# =========================================================

combined_df['is_future'] = 0

combined_df.loc[
    combined_df.index[-len(infer_df):],
    'is_future'
] = 1

# =========================================================
# SORT AGAIN
# =========================================================

combined_df = combined_df.sort_values(
    'capturedAt'
)

# =========================================================
# EXCLUDE NON-FILLABLE
# =========================================================

exclude_cols = [

    'capturedAt',

    'price',
    'price_log1p',
    'log_price',

    'tag1',
    'tag2',

    'is_anchor',
    'is_future'
]

fill_cols = [

    col for col in combined_df.columns
    if col not in exclude_cols
]

# =========================================================
# FORWARD FILL
# PRIORITY:
# 1. EXACT (shopId,itemId,modelId)
# =========================================================

combined_df[fill_cols] = (

    combined_df

    .groupby(
        ['shopId', 'itemId', 'modelId']
    )[fill_cols]

    .ffill()
)

# =========================================================
# FALLBACK:
# itemId
# =========================================================

combined_df[fill_cols] = (

    combined_df

    .groupby(
        ['itemId']
    )[fill_cols]

    .ffill()
)

# =========================================================
# EXTRACT FUTURE
# =========================================================

filled_future_df = combined_df[
    combined_df['is_future'] == 1
].copy()

# =========================================================
# FIX BOOLEAN STRING COLUMNS
# =========================================================

bool_cols = [

    'is_official_shop',
    'is_verified',
    'is_preferred_plus_seller',

    'is_free_shipping',
    'is_pre_order'
]

for col in bool_cols:

    filled_future_df[col] = (

        filled_future_df[col]

        .replace({
            't': 1,
            'f': 0,
            True: 1,
            False: 0
        })

        .fillna(0)

        .astype(np.int8)
    )

# =========================================================
# PREDICT
# =========================================================

X_future = filled_future_df[
    FEATURES
]

pred_log = model.predict(
    X_future
)

filled_future_df['predicted_price'] = np.expm1(
    pred_log
)

# =========================================================
# ANCHOR DF
# =========================================================

anchor_df = filled_future_df[
    filled_future_df['is_anchor'] == 1
].copy()

anchor_df['shift_ratio'] = (

    anchor_df['price']

    /

    anchor_df['predicted_price']
)

# =========================================================
# MODEL-LEVEL DRIFT
# =========================================================

model_drift = (

    anchor_df

    .groupby(
        ['date', 'modelId']
    )['shift_ratio']

    .median()

    .reset_index()

    .rename(
        columns={
            'shift_ratio': 'model_drift_factor'
        }
    )
)

# =========================================================
# ITEM-LEVEL DRIFT
# =========================================================

item_drift = (

    anchor_df

    .groupby(
        ['date', 'itemId']
    )['shift_ratio']

    .median()

    .reset_index()

    .rename(
        columns={
            'shift_ratio': 'item_drift_factor'
        }
    )
)

# =========================================================
# SHOP-LEVEL DRIFT
# =========================================================

shop_drift = (

    anchor_df

    .groupby(
        ['date', 'shopId']
    )['shift_ratio']

    .median()

    .reset_index()

    .rename(
        columns={
            'shift_ratio': 'shop_drift_factor'
        }
    )
)

# =========================================================
# GLOBAL DAILY DRIFT
# =========================================================

global_drift = (

    anchor_df

    .groupby('date')['shift_ratio']

    .median()

    .reset_index()

    .rename(
        columns={
            'shift_ratio': 'global_drift_factor'
        }
    )
)

# =========================================================
# MERGE DRIFT FACTORS
# =========================================================

filled_future_df = filled_future_df.merge(

    model_drift,

    on=['date', 'modelId'],

    how='left'
)

filled_future_df = filled_future_df.merge(

    item_drift,

    on=['date', 'itemId'],

    how='left'
)

filled_future_df = filled_future_df.merge(

    shop_drift,

    on=['date', 'shopId'],

    how='left'
)

filled_future_df = filled_future_df.merge(

    global_drift,

    on='date',

    how='left'
)

# =========================================================
# HIERARCHICAL DRIFT FALLBACK
# model -> item -> shop -> global
# =========================================================

filled_future_df['final_drift_factor'] = (

    filled_future_df['model_drift_factor']

    .fillna(
        filled_future_df['item_drift_factor']
    )

    .fillna(
        filled_future_df['shop_drift_factor']
    )

    .fillna(
        filled_future_df['global_drift_factor']
    )

    .fillna(1.0)
)

# =========================================================
# APPLY DRIFT
# =========================================================

filled_future_df['predicted_price_adjusted'] = (

    filled_future_df['predicted_price']

    *

    filled_future_df['final_drift_factor']
)

# =========================================================
# OPTIONAL ROUNDING
# =========================================================

filled_future_df['predicted_price_adjusted'] = (

    filled_future_df['predicted_price_adjusted']

    .round(0)
)

# =========================================================
# KEEP TRUE ANCHOR PRICE
# =========================================================

filled_future_df.loc[
    filled_future_df['is_anchor'] == 1,
    'predicted_price_adjusted'
] = filled_future_df['price']

# =========================================================
# FINAL OUTPUT
# =========================================================

submission_df = filled_future_df[

    [
        'capturedAt',
        'shopId',
        'itemId',
        'modelId',

        'predicted_price_adjusted'
    ]

].copy()

submission_df = submission_df.rename(
    columns={
        'predicted_price_adjusted': 'predicted_price'
    }
)

# =========================================================
# CHECK
# =========================================================

print("=" * 60)
print("INFERENCE COMPLETE")
print("=" * 60)

print(f"Rows predicted: {len(submission_df):,}")

print("=" * 60)

display(
    submission_df.head()
)
```
